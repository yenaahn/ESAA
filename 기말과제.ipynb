K-Means Clustering 진행하기

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
card_stand=card[["Customer_Age"    ,          
"Dependent_count"  ,
"Months_on_book",
"Total_Relationship_Count",
"Months_Inactive_12_mon",
"Contacts_Count_12_mon",
"Credit_Limit",
"Total_Revolving_Bal"   ,    
"Avg_Open_To_Buy"     ,      
"Total_Amt_Chng_Q4_Q1"  ,     
"Total_Trans_Amt"     ,       
 "Total_Trans_Ct"    ,          
"Total_Ct_Chng_Q4_Q1"  ,       
 "Avg_Utilization_Ratio" ]]
```

```python
card_stand.head()
```

```python
#표준화하기
card_stand_scale=StandardScaler().fit_transform(card_stand)
pd.DataFrame(card_stand_scale)
```

```python
#클러스터링하기
k=4 #변수를 잘 바꿔주기 위해 임의의 변수 설정
model = KMeans(n_clusters = k)
#scaled 된 data에 학습
model.fit(card_stand_scale)
```

```python
#클러스터 결과보기
card['cluster']=model.fit_predict(card_stand_scale)
```

```python
#그룹별 결과보기
card.groupby('cluster').count()
#cluster0>3>1>2 순으로 많이 분포하고 있음을 알 수 있다.
```

```python
#그룹별 평균보기
card.groupby('cluster').mean()
```

```python
#attrition 여부를 알아야 함
card['Attrition_Flag'].astype('category')
card['Attrition_Flag'].replace({'Existing Customer':1,'Attrited Customer':0}, inplace=True)
#지속적인 카드 이용객은 1로, 이탈한 고객은 0으로 표시함
```
